```python
def temperature_converter():
    print("Temperature Converter: Celsius, Fahrenheit, Kelvin")
    print("Enter your choice:")
    print("1. Convert from Celsius")
    print("2. Convert from Fahrenheit")
    print("3. Convert from Kelvin")

    choice = int(input("Choice: "))

    if choice == 1:
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = (celsius * 9/5) + 32
        kelvin = celsius + 273.15
        print(f"{celsius}°C = {fahrenheit}°F")
        print(f"{celsius}°C = {kelvin}K")

    elif choice == 2:
        fahrenheit = float(input("Enter temperature in Fahrenheit: "))
        celsius = (fahrenheit - 32) * 5/9
        kelvin = celsius + 273.15
        print(f"{fahrenheit}°F = {celsius}°C")
        print(f"{fahrenheit}°F = {kelvin}K")

    elif choice == 3:
        kelvin = float(input("Enter temperature in Kelvin: "))
        celsius = kelvin - 273.15
        fahrenheit = (celsius * 9/5) + 32
        print(f"{kelvin}K = {celsius}°C")
        print(f"{kelvin}K = {fahrenheit}°F")

    else:
        print("Invalid choice. Please select 1, 2, or 3.")

temperature_converter()
```

#### Act

```python
def word_count_with_frequency(sentence):
    words = sentence.split()
    word_count = len(words)
    word_frequency = {}

    for word in words:
        word = word.lower()
        word_frequency[word] = word_frequency.get(word, 0) + 1

    return word_count, word_frequency

sentence = input("Enter a sentence: ")
count, frequency = word_count_with_frequency(sentence)
print(f"The number of words in the sentence is: {count}")
print("Word frequencies:", frequency)
```

```python
def merge_dictionaries():
    print("Enter the first dictionary (key-value pairs separated by commas, e.g., a:1,b:2):")
    dict1_input = input("First dictionary: ")
    dict1 = {k.strip(): v.strip() for k, v in (item.split(":") for item in dict1_input.split(","))}
    
    print("Enter the second dictionary (key-value pairs separated by commas, e.g., x:10,y:20):")
    dict2_input = input("Second dictionary: ")
    dict2 = {k.strip(): v.strip() for k, v in (item.split(":") for item in dict2_input.split(","))}
    
    merged_dict = {**dict1, **dict2}
    print("\nMerged Dictionary:")
    print(merged_dict)

merge_dictionaries()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Load the CSV file
df = pd.read_csv("4laptops.csv")
print(df)  # Display the dataframe

# One-Hot Encoding without scikit-learn
df_encoded = pd.get_dummies(df)
print(df_encoded)

# One-Hot Encoding using scikit-learn
enc = OneHotEncoder()

# Fit and transform the 'Category' column
enc_data = enc.fit_transform(df[['Category']])

# Identifying the list of categories considered for encoding
print(enc.categories_)

# Converting the encoded data into an array
enc_array = enc_data.toarray()

# Creating a dataframe without column names
enc_df = pd.DataFrame(enc_array)
print(enc_df)

# Creating a dataframe with proper column names
columns = ["2 in 1 Convertible", "Gaming", "Netbook", "Notebook", "Ultrabook", "Workstation"]
enc_df = pd.DataFrame(enc_array, columns=columns)
print(enc_df)

# Joining original dataframe with encoded dataframe
df_final = df.join(enc_df)
print(df_final)

# Label Encoding
le = LabelEncoder()

# Fit and transform the 'RAM' column
df['RAM'] = le.fit_transform(df['RAM'])
print(df)

# Display label classes
print(le.classes_)
```
